```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
import tensorflow as tf
import numpy as np
```

```python
vectorizer = CountVectorizer(vocabulary=pickle.load(open("fitur.pkl", "rb")))
vectorizer._validate_vocabulary()
transformer = TfidfTransformer(
    norm = 'l2',
    use_idf = True
)
```

```python
user_input = 'Jokowi berniat mengundurkan diri dari jabatan presiden bulan depan'
x = transformer.transform(vectorizer.transform(np.array([user_input]))).toarray()
```

```python
model = tf.keras.models.load_model('verify_model.h5', compile=False)
```

```python
prediksi_hoax = model.predict(x)
prediksi_hoax *= 100
prediksi_hoax
```
